# 1. Holography

## 1.1 History

As early as 1948, Dennis Gabor introduced a method for three-dimensional photography of objects, known as *holography*. This requires coherent light, which only became readily available with the invention of the laser. Holography is used in holographic interferometry when objects scatter light diffusely and can be compared interferometrically with themselves at a later time. Modern cameras can also record holograms digitally and analyze them. [Digital holographic microscopy](https://en.wikipedia.org/wiki/Digital_holographic_microscopy) (DHM) is another important field of application.

## 1.2 Recording a hologram

In photography, the quantity that is recorded is the intensity of light.  For this purpose, objects are usually illuminated with incoherent light.  With this type of recording, however, the phase information of the light is lost. In holography, in addition to the light originating from the object, a reference wave is superimposed so that both the amplitude and the phase of the light are recorded.  
A coherent light source is therefore required. The phase information is stored in the interference pattern.

Reconstruction of the holographic image is achieved by illuminating the developed hologram once again with the reference wave.

Let us now look more closely at the recording of a hologram:

<img src="figures/8.2_1.png" alt="8.2_1.jpg" style="width: 450px;"/>

We denote the wave coming from the object by $S(x,y,t)$ and the reference wave falling directly onto the hologram by $R(x,y,t)$.  
The electric field in the hologram plane $z=0$ is then

$$E(x,y,t)=S(x,y,t) + R(x,y,t)\tag{8.2.1}$$

For monochromatic light, we have $R(x,y,t)=R(x,y)\,e^{-i\omega t}$, and for a stationary object, $S(x,y,t)=S(x,y)\,e^{-i\omega t}$.

The intensity distribution on the holographic plate is therefore

\begin{eqnarray*}
I(x,y)&=&\left|R(x,y)+S(x,y)\right|^2=(R+S)(R+S)^*\\
&=&RR^*+SS^*+R^*S+RS^*\tag{8.2.2}
\end{eqnarray*}

The photographic plate is sensitive to the total energy received per unit area during the exposure time $t_B$.  
It converts this quantity into an optical density and a change in refractive index.  
The exposure is

$$ B(x,y)=\int_0^{t_B} I(x,y,t)\,\mathrm{d}t=\int_0^{t_B} E(x,y,t)E^*(x,y,t)\,\mathrm{d}t$$

As a result, the plate forms a complex transmittance for the electric field amplitude,

$$\tau=\frac{E_t}{E_e}=T(x,y)\,e^{i\varphi(x,y)}$$

<img src="figures/8.2_2.png" alt="8.2_2.jpg" style="width: 200px;"/>

There are two limiting cases, namely the amplitude hologram ($\varphi=\mathrm{const.}$) and the phase hologram ($T=\mathrm{const.}$).  
The advantage of the phase hologram is that it absorbs almost no light and therefore produces a brighter reconstructed image.

For the standard exposure and development process of a holographic plate, the result is an amplitude hologram.  
The figure below shows the typical transmittance $T$ of the developed plate as a function of the exposure $B$:

<img src="figures/8.2_3.png" alt="8.2_3.png" style="width: 300px;"/>

and in an intermediate region around the exposure $B_A$, this can be approximated linearly as

$$T=a-b\,B=a-b\,I\,t_B$$

Thus, the transmittance of an amplitude hologram is

$$T(x,y)=a-b\,t_B\left(R R^* + S S^* + R^* S +R S^*\right)(x,y)\quad.\tag{8.2.3}$$

For a phase hologram, the situation is very similar.  
Again, we choose an operating point $B'_A$ for the exposure and let the phase vary with the exposure according to

<img src="figures/8.2_4.png" alt="8.2_4.png" style="width: 300px;"/>

$$\varphi(I)=a'+b'\,t_B\,I$$

Then the transmittance of the plate is

$$\tau=e^{i\varphi(I)}$$

If the phase satisfies $\varphi(I)\ll\pi/2$, the exponential function can be linearized:

$$\tau=e^{i\varphi(I)}\approx1+i\varphi(I)=(1+ia')+
ib't_B\left(R R^* + S S^* + R^* S +R S^*\right)\tag{8.2.4}$$

Apart from the constants, Eq. (8.2.3) for the amplitude hologram and Eq. (8.2.4) for the phase hologram are therefore very similar.

## 1.3 Reproduction / reconstruction of the hologram

During reproduction (reconstruction) of the hologram, the hologram is illuminated with the same reference wave that was used during recording.  The reference wave $R(x,y)e^{i\omega t}$ is therefore modulated by the complex transmittance $\tau(x,y)$.   If the hologram is an amplitude hologram, then

\begin{eqnarray*}
E_t(x,y)&=&T(x,y)\,E_e\\
&=&R\,a - b\,t_B\,R\left(R R^* + S S^* + R^* S +R S^*\right)\quad .\tag{8.3.1}
\end{eqnarray*}

We obtain 4 terms, which have the following meaning:

1. $(a-b\,t_B|R|^2)R\propto R$ is the reference wave $R$ multiplied by a constant factor (zero-order diffraction term)

2. $b\,t_B\,|S|^2\,R$ is a zero-order diffraction term broadened by the speckle pattern $|S|^2$

3. $b\,t_B\,|R|^2S\propto S$ is the direct image (virtual image)

4. $b\,t_B R^2\,S^*$ is the conjugate image (real image)

### 1.3.1 Position of the images

The position of the images is illustrated in the figure below.  
The direct image is a virtual image located behind the hologram, similar to how an object seen through a window appears to be behind the window.  
The conjugate image is a real image located in front of the hologram and can be observed on a screen.

<img src="figures/8.3_1.png" alt="8.3_1.png" style="width: 500px;"/>

For simplicity, let us assume that the object wave is a plane wave incident on the hologram at an angle $\gamma$.  
The conjugate wave $S^*$ then propagates in the direction

$$S=e^{ikx\sin \gamma}\rightarrow S^*=e^{-ikx\sin\gamma}=e^{ikx\sin(-\gamma)}\quad .\tag{8.3.2}$$

Thus, the wave $S^*$ can be interpreted as a wave leaving the hologram at the angle $-\gamma$.  
Now let the reference wave also be a plane wave, $R=\exp(ikx\sin\alpha)$, incident on the holographic plate at the angle $\alpha$.  
Because the expression contains $R^2$, we have

$$R^2=e^{ikx2\sin\alpha}=e^{ikx\sin\alpha'}\quad .\tag{8.3.3}$$

For small angles, $\alpha'=2\alpha$.  
This is also the situation shown in the figure above.  
Using Eq. (8.3.2) and Eq. (8.3.3), the image term $R^2S^*$ becomes

\begin{eqnarray*}
R^2\,S^*&=&e^{ikx2\sin\alpha}\,e^{i k x \sin(-\gamma)}\\
&=&e^{ikx(2\sin\alpha+\sin(-\gamma))}
\end{eqnarray*}

Now assume that the conjugate image is also a plane wave leaving the hologram at the angle $\beta$, i.e. let $R^2\,S^*=\exp(ikx\sin\beta)$.  
This yields the relation

$$\sin\beta=2\sin\alpha+\sin(-\gamma)\quad .$$

As $\alpha$ or $\gamma$ increase, the conjugate image becomes increasingly distorted, until

$$\sin\beta=2\sin\alpha+\sin(-\gamma)>1$$

and the angle $\beta$ no longer exists.  
In that case, this part of the conjugate image disappears.

## 1.5 Digital hologram

A hologram can also be generated without an optical setup, namely by calculating it algorithmically.  
It is then transferred onto a suitable medium.  
This could be a photographically exposed mask with sufficient resolution.  
Liquid-crystal displays can also be used as phase holograms.

### 1.5.1 Numerical simulation of a hologram

For simplicity, let us assume that an object consists of individual points $P_l,\;l=1,2,\dots,L$ at positions $(x_l,y_l,z_l)$.  
The hologram plate is located in the plane $z=0$ and consists of the pixels
$\{(x_m,y_n,0),\,m=1,\dots,M,\,n=1,\dots,N\}$, as shown in the figure:

<img src="figures/8.5_1.png" alt="8.5_1.png" style="width: 450px;"/>

Each point $P_l$ is a source of a spherical wave, producing at the pixel position $(x_m,y_n)$ on the hologram plate the complex amplitude

$$E_{lmn}=A_l\,e^{i\varphi_l}\,\frac{e^{i\,k\,r_{lmn}}}{r_{lmn}}$$

where $A_l$ and $\varphi_l$ are the amplitude and phase of the spherical wave emitted by point $P_l$.  
$r_{lmn}$ is the distance between this point and the pixel $(x_m,y_n)$ on the hologram plate:

$$r_{lmn}=\sqrt{\left(x_l-x_m\right)^2+\left(y_l-y_n\right)^2+z_l^2}$$

The amplitude of the object wave is the sum of all these waves:

$$E_{Smn}=\sum_{l=1}^L\,E_{lmn}\quad .$$

After superposition with the reference wave $E_R$, the intensity at the point $(x_m,y_n)$ is

$$I_{mn}=\left|E_{Rmn}+E_{Smn}\right|^2=|E_{mn}|^2\quad \tag{8.5.1}.$$

In principle, this intensity can be calculated for every point on the hologram plate.  
However, for a hologram with high resolution and complex geometry, the computational effort becomes very large.  
We therefore introduce several simplifications:

1. All waves emitted by the object points have the same amplitude, i.e. $A_l=\mathrm{const.}$ for all $l$.

2. All object waves have the same phase, i.e. $\varphi_l=0$ for all $l$.

3. If the object is far from the hologram and small in extent, the amplitudes of the object waves can be treated as equal, i.e.

  $$\frac{A_l\,e^{i\varphi_l}}{r_{lmn}}=\frac{A_l}{r_{lmn}}=\mathrm{const.}=1\quad,$$

  where 1 is simply an arbitrary normalization.

4. This leaves only the expression $E_{lmn}=e^{i k r_{lmn}}$, and we set the wave number $k=1$:

  $$E_{lmn}=e^{i\,r_{lmn}}$$

5. The reference wave is assumed to fall perpendicularly onto the hologram plate.  
   Again, we choose a constant phase such that

  $$E_{Rmn}=E_R=\mathrm{const.},\quad\mathrm{for\ all}\;m,n;\; E_R\,\mathrm{real}$$

If, instead of using the grayscale image corresponding to Eq. (8.5.1), one uses a black-and-white pattern, the calculation and production of the hologram can be simplified even further.  
In the figure below, an average intensity $\overline{I}$ is used; values above it are drawn black, and values below it remain bright.

<img src="figures/8.5_2.png" alt="8.5_2.png" style="width: 450px;"/>

We now consider the summation of the object waves $E_{lmn}$ in the complex plane.  
If the intensity of the object waves is much smaller than that of the reference wave, then the average intensity is essentially determined by

$$\overline{I}\approx|E_R|^2=E_R^2$$

The addition of object and reference waves can then be illustrated schematically as follows:

<img src="figures/8.5_3.png" alt="8.5_3.png" style="width: 550px;"/>

If the real part of the sum of the object waves is greater than 0, a black dot is produced; otherwise the point remains white.

Thus, a simple criterion can be given for how a point on the hologram plate is represented:

\begin{eqnarray*}
Re(E_{Smn})>0 &\rightarrow & \mathrm{black\,dot,}\\
\mathrm{else} &\rightarrow & \mathrm{none.}
\end{eqnarray*}

Therefore, we no longer need to write the reference wave explicitly, and only the real part of the sum of object waves matters.  
This real part depends only on the cosine of the distances $r_{lmn}$:

$$
Re(E_{Smn})=\sum_{i=1}^L\cos(r_{lmn})
$$

The program below can be used to generate a digital (binary) hologram from point sources.  
You can use it in the exercises.

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image
nx=640
ny=480
#Points

npts=1
phi=np.linspace(0,2*np.pi,npts)

P=np.zeros((npts,3))
P[0]=[nx/2,ny/2,-200]

H=np.zeros((ny,nx))
dyv=np.zeros(npts)

for j in range(ny):
    
    for i in range(npts):
        dyv[i]=(P[i][1]-j)**2
    
    for k in range(nx):
        sum=0.
        for i in range(npts):
            dx=P[i,0]-k
            sum += np.cos((dx*dx+dyv[i]+P[i,2]*P[i,2])**.5)
        
        if (sum>=0):
            H[j,k]=1.
            
plt.figure(figsize=(10,10))
plt.imshow(H,cmap='Greys', interpolation='none',aspect='equal')
image.imsave('test.png', H, cmap='gray')
```
