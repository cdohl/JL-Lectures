# 1. 全息术

## 1.1 历史

早在 1948 年，Dennis Gabor 就提出了一种记录物体三维图像的方法，这就是 *全息术*。这需要相干光，而相干光直到激光发明之后才变得容易获得。全息术被用于全息干涉测量：当物体对光产生漫散射，并且能够在稍后的时刻与其自身进行干涉比较时，就可以应用这种方法。现代相机也能够对全息图进行数字记录与分析。[数字全息显微术](https://en.wikipedia.org/wiki/Digital_holographic_microscopy)（DHM）也是一个重要的应用领域。

## 1.2 全息图的记录

在摄影中，所记录的是光的强度。为此，通常用非相干光照明物体。但采用这种记录方式时，光的相位信息会丢失。在全息术中，除了来自物体的光之外，还叠加一个参考波，从而能够同时记录光的振幅和相位。  
因此必须使用相干光源。相位信息被储存在干涉图样中。

全息图像的重建是通过再次用参考波照射已显影的全息图来实现的。

现在我们更仔细地来看看全息图的记录过程：

<img src="figures/8.2_1.png" alt="8.2_1.jpg" style="width: 450px;"/>

我们把来自物体的波记作 $S(x,y,t)$，把直接照射到全息图上的参考波记作 $R(x,y,t)$。  
因此，在全息图平面 $z=0$ 上的电场为

$$E(x,y,t)=S(x,y,t) + R(x,y,t)\tag{1.2.1}$$

对于单色光，有 $R(x,y,t)=R(x,y)\,e^{-i\omega t}$，而对于静止物体，有 $S(x,y,t)=S(x,y)\,e^{-i\omega t}$。

于是，全息底片上的强度分布为

\begin{eqnarray*}
I(x,y)&=&\left|R(x,y)+S(x,y)\right|^2=(R+S)(R+S)^*\\
&=&RR^*+SS^*+R^*S+RS^*\tag{1.2.2}
\end{eqnarray*}

照相底片对曝光时间 $t_B$ 内单位面积接收到的总能量敏感。  
底片会把这一物理量转化为光学密度以及折射率的变化。  
曝光量为

$$ B(x,y)=\int_0^{t_B} I(x,y,t)\,\mathrm{d}t=\int_0^{t_B} E(x,y,t)E^*(x,y,t)\,\mathrm{d}t$$

由此，底片对电场振幅形成一个复透射率

$$\tau=\frac{E_t}{E_e}=T(x,y)\,e^{i\varphi(x,y)}$$

<img src="figures/8.2_2.png" alt="8.2_2.jpg" style="width: 200px;"/>

存在两个极限情形，即振幅全息图（$\varphi=\mathrm{const.}$）和相位全息图（$T=\mathrm{const.}$）。  
相位全息图的优点在于它几乎不吸收光，因此在重建时能够产生更明亮的图像。

对于全息底片的标准曝光和显影过程，所得到的是振幅全息图。  
下图给出了已显影底片的典型透射率 $T$ 随曝光量 $B$ 的变化关系：

<img src="figures/8.2_3.png" alt="8.2_3.png" style="width: 300px;"/>

并且在曝光量 $B_A$ 附近的一个中间区域内，可以用线性关系近似：

$$T=a-b\,B=a-b\,I\,t_B$$

因此，振幅全息图的透射率为

$$T(x,y)=a-b\,t_B\left(R R^* + S S^* + R^* S +R S^*\right)(x,y)\quad.\tag{1.2.3}$$

对于相位全息图，情况非常类似。  
同样地，我们取一个工作点曝光量 $B'_A$，并令相位随曝光量的变化为

<img src="figures/8.2_4.png" alt="8.2_4.png" style="width: 300px;"/>

$$\varphi(I)=a'+b'\,t_B\,I$$

此时底片的透射率为

$$\tau=e^{i\varphi(I)}$$

当相位满足 $\varphi(I)\ll\pi/2$ 时，可以将指数函数线性化：

$$\tau=e^{i\varphi(I)}\approx1+i\varphi(I)=(1+ia')+
ib't_B\left(R R^* + S S^* + R^* S +R S^*\right)\tag{1.2.4}$$

因此，除了常数不同之外，振幅全息图的式 (1.2.3) 与相位全息图的式 (1.2.4) 非常相似。

## 1.3 全息图的再现 / 重建

在全息图的再现（重建）过程中，全息图要用与记录时相同的参考波照射。  
因此，参考波 $R(x,y)e^{i\omega t}$ 会被复透射率 $\tau(x,y)$ 调制。  
如果该全息图是振幅全息图，则有

\begin{eqnarray*}
E_t(x,y)&=&T(x,y)\,E_e\\
&=&R\,a - b\,t_B\,R\left(R R^* + S S^* + R^* S +R S^*\right)\quad .\tag{1.3.1}
\end{eqnarray*}

我们得到 4 个项，它们分别具有如下含义：

1. $(a-b\,t_B|R|^2)R\propto R$ 是参考波 $R$ 乘以一个常数因子（零级衍射项）

2. $b\,t_B\,|S|^2\,R$ 是被散斑图样 $|S|^2$ 展宽的零级衍射项

3. $b\,t_B\,|R|^2S\propto S$ 是直接像（虚像）

4. $b\,t_B R^2\,S^*$ 是共轭像（实像）

### 1.3.1 图像的位置

图像的位置如下图所示。  
直接像是一个虚像，它位于全息图之后，类似于透过窗户看到窗后物体的情形。  
共轭像是实像，位于全息图前方，可以在屏幕上观察到。

<img src="figures/8.3_1.png" alt="8.3_1.png" style="width: 500px;"/>

为简单起见，设物波是一束以角度 $\gamma$ 入射到全息图上的平面波。  
则共轭波 $S^*$ 的传播方向满足

$$S=e^{ikx\sin \gamma}\rightarrow S^*=e^{-ikx\sin\gamma}=e^{ikx\sin(-\gamma)}\quad .\tag{1.3.2}$$

因此，可以把波 $S^*$ 看作是一束以角度 $-\gamma$ 离开全息图的波。  
现在再令参考波也是平面波，即 $R=\exp(ikx\sin\alpha)$，它以角度 $\alpha$ 入射到全息底片上。  
由于表达式中出现了 $R^2$，因此有

$$R^2=e^{ikx2\sin\alpha}=e^{ikx\sin\alpha'}\quad .\tag{1.3.3}$$

对于小角度，有 $\alpha'=2\alpha$。  
这也是上图中所示的情形。  
由式 (1.3.2) 和式 (1.3.3) 可得，图像项 $R^2S^*$ 为

\begin{eqnarray*}
R^2\,S^*&=&e^{ikx2\sin\alpha}\,e^{i k x \sin(-\gamma)}\\
&=&e^{ikx(2\sin\alpha+\sin(-\gamma))}
\end{eqnarray*}

现在再假设共轭像也是一束以角度 $\beta$ 离开全息图的平面波，也就是说令 $R^2\,S^*=\exp(ikx\sin\beta)$。  
于是得到关系

$$\sin\beta=2\sin\alpha+\sin(-\gamma)\quad .$$

随着 $\alpha$ 或 $\gamma$ 的增大，共轭像会越来越严重地发生畸变，直到

$$\sin\beta=2\sin\alpha+\sin(-\gamma)>1$$

这时角度 $\beta$ 不再存在。  
于是，共轭像的这一部分就会消失。

## 1.5 数字全息图

全息图也可以不借助光学装置，而是通过算法计算得到。随后再将其转移到合适的介质上。这可以是一张具有足够分辨率的照相曝光掩模。液晶显示器也可以作为相位全息图使用。

### 1.5.1 全息图的数值模拟

为简单起见，设一个物体由若干离散点 $P_l,\;l=1,2,\dots,L$ 组成，这些点位于 $(x_l,y_l,z_l)$。  
全息板位于平面 $z=0$ 上，由像素点
$\{(x_m,y_n,0),\,m=1,\dots,M,\,n=1,\dots,N\}$ 构成，如下图所示：

<img src="figures/8.5_1.png" alt="8.5_1.png" style="width: 450px;"/>

每个点 $P_l$ 都是一个球面波源，它会在全息板上的像素位置 $(x_m,y_n)$ 处产生一个复振幅

$$E_{lmn}=A_l\,e^{i\varphi_l}\,\frac{e^{i\,k\,r_{lmn}}}{r_{lmn}}$$

其中，$A_l$ 和 $\varphi_l$ 分别是由点 $P_l$ 发出的球面波的振幅和相位。  
$r_{lmn}$ 是该点到全息板像素点 $(x_m,y_n)$ 的距离：

$$r_{lmn}=\sqrt{\left(x_l-x_m\right)^2+\left(y_l-y_n\right)^2+z_l^2}$$

物光波的振幅就是所有这些波的总和：

$$E_{Smn}=\sum_{l=1}^L\,E_{lmn}\quad .$$

再与参考波 $E_R$ 叠加后，在点 $(x_m,y_n)$ 处的强度为

$$I_{mn}=\left|E_{Rmn}+E_{Smn}\right|^2=|E_{mn}|^2\quad \tag{1.5.1}.$$

原则上，可以计算全息板上每一个点处的这一强度。  
但是，如果要计算具有高分辨率和复杂几何结构的全息图，计算量会非常大。  
因此我们引入若干简化：

1. 每个物点发出的光波具有相同的振幅，即对所有 $l$ 有 $A_l=\mathrm{const.}$。

2. 每个物波具有相同的相位，即对所有 $l$ 有 $\varphi_l=0$。

3. 如果物体距离全息图很远并且尺寸较小，则可以把各物波的振幅视为相同，即

  $$\frac{A_l\,e^{i\varphi_l}}{r_{lmn}}=\frac{A_l}{r_{lmn}}=\mathrm{const.}=1\quad,$$

  其中 1 只是任意选取的归一化值。

4. 这样就只剩下表达式 $E_{lmn}=e^{i k r_{lmn}}$，并取波数 $k=1$：

  $$E_{lmn}=e^{i\,r_{lmn}}$$

5. 参考波垂直照射到全息板上。  
   同样地，我们取一个常数相位，使得

  $$E_{Rmn}=E_R=\mathrm{const.},\quad\mathrm{对于所有}\;m,n;\; E_R\,\mathrm{为实数}$$

如果不用式 (1.5.1) 所对应的灰度图，而改用黑白图样，那么全息图的计算和制作还可以进一步简化。  
在下图中，使用了一个平均强度 $\overline{I}$；高于该值的部分画成黑色，低于该值的部分保持明亮。

<img src="figures/8.5_2.png" alt="8.5_2.png" style="width: 450px;"/>

现在我们在复平面中考察物波 $E_{lmn}$ 的求和。  
如果物波的强度远小于参考波的强度，则平均强度基本上由下式决定：

$$\overline{I}\approx|E_R|^2=E_R^2$$

此时，物波与参考波的叠加可以用下图示意表示：

<img src="figures/8.5_3.png" alt="8.5_3.png" style="width: 550px;"/>

如果物波总和的实部大于 0，则得到一个黑点；否则该点保持白色。

因此，可以给出一个简单的判据，用来决定全息板上的某一点如何表示：

\begin{eqnarray*}
Re(E_{Smn})>0 &\rightarrow & \mathrm{black\,dot,}\\
\mathrm{else} &\rightarrow & \mathrm{none.}
\end{eqnarray*}

因此，我们不再需要显式写出参考波，而只需要关心物波总和的实部。  
而这一实部只取决于距离 $r_{lmn}$ 的余弦：

$$
Re(E_{Smn})=\sum_{i=1}^L\cos(r_{lmn})
$$

下面的程序可用于生成由点源构成的数字（二值）全息图。  
你可以在练习中使用它。

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image
nx=640
ny=480
#Points

npts=1
phi=np.linspace(0,2*np.pi,npts)

P=np.zeros((npts,3))
P[0]=[nx/2,ny/2,-200]

H=np.zeros((ny,nx))
dyv=np.zeros(npts)

for j in range(ny):
    
    for i in range(npts):
        dyv[i]=(P[i][1]-j)**2
    
    for k in range(nx):
        sum=0.
        for i in range(npts):
            dx=P[i,0]-k
            sum += np.cos((dx*dx+dyv[i]+P[i,2]*P[i,2])**.5)
        
        if (sum>=0):
            H[j,k]=1.
            
plt.figure(figsize=(10,10))
plt.imshow(H,cmap='Greys', interpolation='none',aspect='equal')
image.imsave('holo1.png', H, cmap='gray')
```

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import image
nx=640
ny=480
#Points
R=200
npts=20
phi=np.linspace(0,9./10.*2*np.pi,npts)
za=np.linspace(-300,-100,npts)
P=np.zeros((100,3))
for i in range(npts):
    P[i]=[nx/2+R*np.sin(phi[i]),ny/2+R*np.cos(phi[i]),-300]

H=np.zeros((ny,nx))
dyv=np.zeros(npts)

for j in range(ny):
    
    for i in range(npts):
        dyv[i]=(P[i][1]-j)**2
    
    for k in range(nx):
        sum=0.
        for i in range(npts):
            dx=P[i,0]-k
            sum += np.cos((dx*dx+dyv[i]+P[i,2]*P[i,2])**.5)
        
        if (sum>=0):
            H[j,k]=1.
            
plt.figure(figsize=(8,8))
plt.imshow(H,cmap='Greys', interpolation='none',aspect='equal');
image.imsave('holo2.png', H, cmap='gray')
```
